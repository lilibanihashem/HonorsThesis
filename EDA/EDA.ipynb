```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=3, cols=3,vertical_spacing = 0.15, subplot_titles = titles
)

y = 1
for i in ["WF9_CHAR_RACE", "WF9_CHAR_HISP","WF9_CHAR_COUNTRY_BORN", "WF9_CHAR_HHINCOME","WF9_CAREER_UNION", "WF9_CHAR_MARITAL", "WF9_CHAR_GOVT_PRGM", 
          "WF9_CHAR_HEALTH_INSRNCE", "WF9_CAREER_EXPERIENCE"
                                                            
                                           ]:
    crosstb = pd.crosstab(data[i],data["WF9_CESD7_CUT"], margins=True, margins_name="Total")
    names = labels[y-1]
    
    crosstb.index = names


    crosstb[1] = crosstb[1]/crosstb["Total"]
    
    if y%3 == 0:
        col = 3
    else: 
        col = y%3
    
    fig.add_trace(
    go.Bar(x=names, y=crosstb[1]),
    
    row=(y+2)//3, col= col
    )
    y+=1

fig.update_layout(height=1000, width=1000, title_text="Depression by Demographic Variables", showlegend = False, colorway = ["#1f77b4", "#0b0754"])
fig.update_xaxes(tickangle=325)


fig.show()
```
